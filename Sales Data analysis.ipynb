```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
data=pd.read_csv('/content/Sales Data.csv')
data.head()
```

```python
data.dtypes
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data=data.dropna(how='all')
data.shape
```

```python
'04/19/19 08:46'.split('/')[0]
```

```python
def month(x):
    if isinstance(x, str):
        return x.split('/')[0]
    else:
        return None
```

```python
data['Month']=data['Order Date'].apply(month)
```

```python
data.dtypes
```

```python
data['Month'].unique()
```

```python
filter=data['Month']=='Order Date'
len(data[~filter])
```

```python
data=data[~filter]
```

```python
data.shape
```

```python
data.head()
```

```python
data['Month']
```

```python
data.dtypes
```

```python
data['Price Each']=data['Price Each'].astype(float)
```

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'], errors='coerce')
data['Quantity Ordered'] = data['Quantity Ordered'].fillna(0).astype(int)
```

```python
data['Sales']=data['Quantity Ordered']*data['Price Each']
data.head(5)
```

```python
data.groupby('Month')['Sales'].sum()
```

```python
'917 1st St, Dallas, TX 75001'.split(',')[1]
```

```python
def city(address):
    if isinstance(address, str):
        return address.split(',')[0]
    else:
        return None
```

```python
data['city']=data['Purchase Address'].apply(city)
```

```python
data.groupby('city')['city'].count()
```

```python
plt.bar(data.groupby('city').size().index, data.groupby('city').size())
plt.xticks(rotation='horizontal')
plt.ylabel('received orders')
plt.xlabel('city names')
plt.show()
```

```python
data['Hour']=pd.to_datetime(data['Order Date']).dt.hour
```

```python
keys=[]
hour=[]
for key,hour_df in data.groupby('Hour'):
  keys.append(key)
  hour.append(len(hour_df))
```

```python
plt.grid()
plt.plot(keys,hour)
```

```python
data.groupby('Product')['Quantity Ordered'].sum().plot(kind='bar')
```

```python
data.groupby('Product')['Quantity Ordered'].sum().plot(kind='bar')
```

```python
products=data.groupby('Product')['Quantity Ordered'].sum().index
quantity=data.groupby('Product')['Quantity Ordered'].sum()
price=data.groupby('Product')['Price Each'].mean()
```

```python
plt.figure(figsize=(40,24))
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,quantity,color='g')
ax2.plot(products,price,'b-')
ax1.set_xticklabels(products,rotation='vertical',size=8)
```
